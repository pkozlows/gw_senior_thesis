```python
from sympy import symbols, diff, ln, latex, solve

# Define symbols
m, x, alpha, mu = symbols('m x alpha mu')

# Define the expression for g
g = -1/2 * alpha * (m * x)**2 - mu * x + ((1 + m)/2)*x*ln(((1 + m)/2)*x) + ((1 - m)/2)*x*ln(((1 - m)/2)*x) + (1 - x)*ln(1 - x)

# Calculate the first derivatives
dg_dm = diff(g, m)
dg_dx = diff(g, x)

# Solve the equation dg_dx = 0 for mu to get mu as a function of m, x, and alpha
mu_solution = solve(dg_dx, mu)

print(latex(mu_solution))

```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_m_vs_x(alpha_values):
    x_values = np.linspace(0, 1, 100)  # Define the range of x values
    plt.figure(figsize=(12, 8))  # Set the figure size
    
    for alpha in alpha_values:
        # Simplified assumption for demonstration purposes
        m_values = 2 * np.tanh(alpha * x_values)  # Calculate m values based on the tanh function
        
        plt.plot(x_values, m_values, label=f'={alpha}')  # Plot m vs x for each alpha

    plt.title('Plot of $m$ vs $x$ for $\\alpha=\\{1,2, \\ldots, 9\\}$')  # Title
    plt.xlabel('$x$')  # x-axis label
    plt.ylabel('$m$')  # y-axis label
    plt.legend()  # Show legend
    plt.grid(True)  # Show grid
    plt.savefig('m_vs_x.png')  # Save the plot as a PNG file

plot_m_vs_x(range(1, 10))  # Call the function with alpha values from 1 to 9
```
